Write a NumPy program to save two given arrays into a single file in compressed format (.npz format)
and load it.


```python
a  = np.array([1,2,3,4,5])
b = np.array([11,12,13,14,15])
print("Original arrays:")
print(a)
print(b)
np.savez('temp_arr.npz', a_tmp=a, b_tmp=b)
print("Load arrays from the 'temp_arr.npz' file:")
with np.load('temp_arr.npz') as data:
    aa = data['a_tmp']
    bb = data['b_tmp']
    print(aa)
    print(bb)
```

Write a NumPy program to extract all numbers from a given array which are less and greater than a
specified number.

```python
a = np.arange(0,100)
n1 = 35
a_ext1 = a[(a > n1) ]
a_ext2 = a[(a < n1) ]
print ("Original array = \n", a)
print("Elements less than 35 = \n", a_ext1)
print("Elements greater than 35 = \n", a_ext2)
```
