```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn  as sns
```

Write a Python program to draw a line with suitable label in the x axis,
y axis and a title.

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y)
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.title('Line Plot with Labels')
plt.show()
```

Write a Python program to plot two or more lines on same plot with
suitable legends of each line.

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [3, 6, 9, 12, 15]

plt.plot(x,y1, label='Line1')
plt.plot(x,y2, label='Line2')

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Two Lines Plot')

plt.legend()

plt.show()
```

Write a Python program to plot two or more lines with legends, different
widths and colors.


```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [3, 6, 9, 12, 15]

plt.plot(x,y1, label='Line1', color = 'blue', linewidth=2)
plt.plot(x,y2, label='Line2', color = 'red', linewidth=3)

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Two Lines Plot with Legends, Different Widths, and Colors')

plt.legend()

plt.show()
```

Write a Python program to plot two or more lines with different styles.


```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [3, 6, 9, 12, 15]

plt.plot(x,y1, label='Line1', color = 'blue', linestyle = '--')
plt.plot(x,y2, label='Line2', color = 'red')

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Two Lines Plot with Different Styles')

plt.legend()

plt.show()
```

Write a Python program to plot two or more lines and set the line
markers.

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [3, 6, 9, 12, 15]

plt.plot(x,y1, label='Line1', color = 'blue', linestyle = '--', marker = 'o')
plt.plot(x,y2, label='Line2', color = 'red', linestyle = ':', marker = 's')

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Two Lines Plot with Markers')

plt.legend()

plt.show()
```

Write a Python programming to display a bar chart of the popularity of
programming Languages.

Sample data:

Programming languages: Java, Python, PHP, JavaScript, C#, C++

Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7


```python

languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

# Plot the bar chart
plt.bar(languages, popularity)

# Add labels and title
plt.xlabel('Programming Languages')
plt.ylabel('Popularity (%)')
plt.title('Popularity of Programming Languages')

# Display the plot
plt.show()
```

Write a Python programming to display a horizontal bar chart of the
popularity of programming Languages.

Sample data:

Programming languages: Java, Python, PHP, JavaScript, C#, C++

Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python


# Sample data
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

# Plot the horizontal bar chart
plt.barh(languages, popularity)

# Add labels and title
plt.xlabel('Popularity (%)')
plt.ylabel('Programming Languages')
plt.title('Popularity of Programming Languages')

# Display the plot
plt.show()
```

. Write a Python program to create bar plot of scores by group and
gender. Use multiple X values on the same chart for men and women.

```python

groups = ['Group 1', 'Group 2', 'Group 3', 'Group 4', 'Group 5', 'Group 1', 'Group 2', 'Group 3', 'Group 4', 'Group 5']
means = [22, 30, 35, 35, 26, 25, 32, 30, 35, 29]
gender  = ['Men', 'Men', 'Men', 'Men', 'Men', 'Women', 'Women', 'Women', 'Women', 'Women']
data = {
    'Groups' : groups,
    'Means': means,
    'Gender': gender
}
df = pd.DataFrame(data)
sns.barplot(data=df, x='Groups', y='Means', hue='Gender')

# Add title
plt.title('Scores by Group and Gender')

# Show the plot
plt.show()
```
