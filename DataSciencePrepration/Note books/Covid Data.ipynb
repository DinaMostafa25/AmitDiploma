```python
def is_dead(date):
    if date == '9999-99-99':
        return 0
    else :
        return 1
```

```python
df['dead'] = df['DATE_DIED'].apply(is_dead)
```

```python
plt.hist(df['dead']);
```

```python
df.groupby('dead')['CLASIFFICATION_FINAL'].value_counts(normalize= True)
```
