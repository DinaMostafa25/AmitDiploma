```python
import pandas  as pd
import matplotlib.pyplot as  plt
import seaborn as sns
```

```python
df= pd.read_csv("US Police shootings in from 2015-22.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()/len(df)
```

```python
df.describe()
```

```python
df['date'] = pd.to_datetime(df['date'], format = 'mixed')
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data = df, x = 'age', bins = 20, kde = True);
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='manner_of_death')
plt.xlabel('Manner of Death')
plt.ylabel('Count')
plt.title('Distribution of Manner of Death')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='race')
plt.xlabel('Race')
plt.ylabel('Count')
plt.title('Distribution of Race')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='manner_of_death', y='age')
plt.xlabel('Manner of Death')
plt.ylabel('Age')
plt.title('Age vs. Manner of Death')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='race', hue='manner_of_death')
plt.xlabel('Race')
plt.ylabel('Count')
plt.title('Race vs. Manner of Death')
plt.xticks(rotation=45)
plt.legend(title='Manner of Death')
plt.show()
```

```python
df['is_armed'] = df['armed'].apply(lambda x: 1 if x!='unarmed' else 0)
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='is_armed', hue='gender')
plt.xlabel('Armed Status')
plt.ylabel('Count')
plt.title('Armed vs. Gender')
plt.xticks([0, 1], ['Unarmed', 'Armed'])
plt.legend(title='Gender')
plt.show()
```
