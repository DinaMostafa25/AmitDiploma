```python
import pandas as pd
from sklearn.model_selection import train_test_split 
from sklearn.metrics import mean_squared_error, r2_score
import xgboost as xgb
from xgboost import XGBRegressor
```

```python
df = pd.read_csv('train (2).csv')
pd.options.display.max_columns= None
df.head(20)
```

```python
df.drop(columns =  ['PoolQC','MiscFeature','Alley','Fence' ,'MasVnrType', 'FireplaceQu',  'LotFrontage'], inplace = True)
```

```python
X = df.drop(columns = 'SalePrice')
y = df['SalePrice']
```

```python
# Identify categorical columns
categorical_cols = X.select_dtypes(include =['object']).columns
```

```python
# Convert categorical columns to 'Categorical' type
X[categorical_cols] = X[categorical_cols].astype('category')
X_dmatrix = xgb.DMatrix(X, label = y, enable_categorical = True)
# Convert DMatrix to a NumPy array
X_array =X_dmatrix.get_data() 
# Get labels from DMatrix
y_array = X_dmatrix.get_label()
#split data for train and test
# Split the data
train_X, test_X, train_y, test_y = train_test_split(X_array, y_array, test_size=0.25)
```

```python
# Create and fit the XGBRegressor model
my_model = XGBRegressor()
my_model.fit(train_X, train_y)
```

```python
# Make predictions
predictions = my_model.predict(test_X)
```

```python
# Evaluate the model
mse = mean_squared_error(test_y, predictions)
print(f"MSE: {mse:.3f}")

r2= r2_score(test_y, predictions)
print(f"R2 score: {r2:.3f}")
```

```python
xgb_model = XGBRegressor(n_estimators = 1000)
xgb_model.fit(train_X, train_y, early_stopping_rounds =5, eval_set = [(test_X, test_y)])
```

```python
y_pred = xgb_model.predict(test_X)
```

```python
# Evaluate the model
mse = mean_squared_error(test_y, y_pred)
print(f"MSE: {mse:.3f}")

r2= r2_score(test_y, y_pred)
print(f"R2 score: {r2:.3f}")
```

```python
xgb_model = XGBRegressor(n_estimators = 1000, learning_rate = .05)
xgb_model.fit(train_X, train_y, early_stopping_rounds =5, eval_set = [(test_X, test_y)])
```

```python
y_pred = xgb_model.predict(test_X)
```

```python
# Evaluate the model
mse = mean_squared_error(test_y, y_pred)
print(f"MSE: {mse:.3f}")

r2= r2_score(test_y, y_pred)
print(f"R2 score: {r2:.3f}")
```

```python
xgb_model = XGBRegressor(n_estimators = 1000, learning_rate = .05, n_jobs =-1)
xgb_model.fit(train_X, train_y, early_stopping_rounds =5, eval_set = [(test_X, test_y)])
```

```python
y_pred = xgb_model.predict(test_X)
```

```python
# Evaluate the model
mse = mean_squared_error(test_y, y_pred)
print(f"MSE: {mse:.3f}")

r2= r2_score(test_y, y_pred)
print(f"R2 score: {r2:.3f}")
```
